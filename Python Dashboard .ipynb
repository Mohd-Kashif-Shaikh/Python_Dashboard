```python
df=df.fillna(0)
```

```python
df["GDP/Capita"] = np.where(df["population"]!=0, df["gdp"]/df["population"], 0)
```

```python
df["GDP/Capita"]
```

```python
idf = df.interactive()
```

```python
Year_Slider= pn.widgets.IntSlider(name="Year Slider", start=1750, end=2020, step=5, value=1850)
Year_Slider           
```

```python
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name="Y axis",
    options=["co2", "co2_per_capita"],
    button_type="success"
)
yaxis_co2
```

```python
continents=["world","Asia","Europe","North America","South America","Oceania","Africa","Antartica"]

co2_Pipeline=(
    idf[
        (idf.year <= Year_Slider) &
        (idf.country.isin(continents))
    ]
    .groupby(["country","year"])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True) 
)
co2_Pipeline
```

```python
co2_plot = co2_Pipeline.hvplot(x="year",by="country",y=yaxis_co2,line_width=4,title="CO2 emission by continent")
co2_plot
```

```python
co2_table=co2_Pipeline.pipe(pn.widgets.Tabulator,pagination="remote",page_size=10,sizing_mode="stretch_width") 
co2_table
```

```python
co2_vs_Gdp_Pipeline=(
    idf[
        (idf.year == Year_Slider) &
        (~(idf.country.isin(continents)))
    ]
    .groupby(['country','year','GDP/Capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
co2_vs_Gdp_Pipeline
```

```python
co2_vs_Gdp=co2_vs_Gdp_Pipeline.hvplot(x='GDP/Capita',
                                      y='co2', 
                                      by='country', 
                                      size=80, kind="scatter", 
                                      alpha=0.7,
                                      legend=False,
                                      height=400, 
                                      width=300)
co2_vs_Gdp                                                                            
```

```python
yaxis_co2_sources = pn.widgets.RadioButtonGroup(
    name="Y_axis", 
    options=["coal_co2", "oil_co2", "gas_co2"], 
    button_type="success"
)
```

```python
continents_excl_world = ["Asia","Europe","North America","South America","Oceania","Africa","Antartica"]

co2_sources_Bar_Pipeline=(
    idf[
        (idf.year == Year_Slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(["year","country"])[yaxis_co2_sources].sum()
    .to_frame()
    .reset_index()
    .sort_values(by="year")  
    .reset_index(drop=True)
) 
```

```python
co2_sources_Bar_Plot = co2_sources_Bar_Pipeline.hvplot(kind="bar",
                                                       x="country", 
                                                       y=yaxis_co2_sources, 
                                                       title="CO2 source by continent",
                                                       height=400,
                                                       width=400)
co2_sources_Bar_Plot
```

```python
template = pn.template.FastListTemplate(
    title='Global Co2 Emissions Monitoring Dashboard', 
    sidebar=[pn.pane.Markdown("## CO2 Emissions and Their impact on Climate Change"),
             pn.pane.PNG('https://static.vecteezy.com/system/resources/thumbnails/022/685/370/small/climate-change-illustration-with-transparent-background-global-warming-illustration-artwork-ai-generated-png.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"), 
             Year_Slider],
    main=[pn.Row(pn.Column(yaxis_co2,
                           co2_plot.panel(width=700),margin=(0,25)),
                 co2_table.panel(width=500)),
          pn.Row(pn.Column(co2_vs_Gdp.panel(width=600), margin=(0,25)),
                 pn.Column(yaxis_co2_sources,co2_sources_Bar_Plot.panel(width=500)))],
)
template.show()
```
